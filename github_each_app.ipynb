```python
import json
import os
from sentence_transformers import SentenceTransformer, util
import pandas as pd
```

```python
df_reviews = pd.read_csv('https://raw.githubusercontent.com/zeeshansalim1234/Summer2021/main/app_data/App944.csv')
df_commits=pd.read_csv('https://raw.githubusercontent.com/zeeshansalim1234/Summer2021/main/app_data/App944%20(1).csv')
df_reviews.columns = ['A', 'B', 'C', 'D','E','F']
df_commits.columns = ['A', 'B', 'C', 'D','E','F','G']
```

```python
print(df_reviews.C.values)
```

```python
reviews = df_reviews.C.values
commits= df_commits.C.values
commits = list(dict.fromkeys(commits)) #list
print(commits)
```

```python
model = SentenceTransformer('allenai-specter')
review_texts = [review for review in reviews]
corpus_embeddings = model.encode(review_texts, convert_to_tensor=True)
```

```python
date_reviews=df_reviews.E.values
date_commits=df_commits.D.values
```

```python
from termcolor import colored



def search_papers(title):

  result=[]
  result.append("Commit : "+title)
  query_embedding = model.encode(title+'[SEP]', convert_to_tensor=True)

  count = 0

  search_hits = util.semantic_search(query_embedding, corpus_embeddings)
  search_hits = search_hits[0]  #Get the hits for the first query

  #print("Query:", title)
  #print("\n Relevant Reviews:")
  for hit in search_hits:
    
    if(search_hits[count]['score']>=0.85):

      date_review=date_reviews[hit['corpus_id']].strip().split("/")
      date_commit=date_commits[hit['corpus_id']].strip().split("-")

      if(int(date_review[2])<=int(date_commit[0])):
        related_reviews = reviews[hit['corpus_id']]
        print()
        subsetDataFrame = df_reviews[df_reviews['C']== related_reviews]
        k=subsetDataFrame.values
        #print("\n"+str(count)+") "+colored(str(k[0][0]), 'red'))
        #print("similiarity score of " + str(format(search_hits[count-1]['score'],".2f")))
        result.append(str(k[0][2])+" ("+str(format(search_hits[count-1]['score'],".2f"))+")")

    count+=1
      
  return result
   
search_papers("Fixing crash caused by network connectivity problems")

#Fixing crash caused by network connectivity problems
# Adding ability to add or remove a URL from the adblock whitelist
```

```python
result=[]

for i in range(0,len(commits)):
  result.append(search_papers(commits[i]))
```
